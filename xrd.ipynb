<h3>Import des différents modules et récupération de la structure</h3>

```python
from numpy import *
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction import xrd
from jupyter_jsmol.pymatgen import quick_view
from ipywidgets import Layout, widgets, interact

with MPRester("VmJsz9RpLnefKgXu") as m: 
    structure = m.get_structure_by_material_id("mp-567636")
    structure.to(filename="VFeSb.cif")
```

```python
view = quick_view(structure, conventional=True)
display(view)
```

<h3>Calcul du diffractorgramme</h3>

```python
diff = xrd.XRDCalculator("CuKa1")
pics = diff.get_pattern(structure)
```

<h3>Premier pic du diffractogramme</h3>
L'angle entre le rayon incident et le rayon diffracté vaut 2 $\theta$ avec $\theta$ qui est l'angle d'incidence par rapport à la normale du plan.

```python
print("Angle 2θ entre les 2 rayons :{:2f}".format(pics.x[0]))
print("Intensité :{:2f}".format(pics.y[0]))
print("Indices hkl associés :{}".format(pics.hkls[0]))
print("Distance associée :{:2f}".format(pics.d_hkls[0]))
```

<h3>Deuxième pic du diffractogramme</h3>

```python
print("Angle 2θ entre les 2 rayons :{:2f}".format(pics.x[1]))
print("Intensité :{:2f}".format(pics.y[1]))
print("Indices hkl associés :{}".format(pics.hkls[1]))
print("Distance associée :{:2f}".format(pics.d_hkls[1]))
```
